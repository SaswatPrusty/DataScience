```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
amazon_data = pd.read_csv("amazon_data.csv")
```

```python
amazon_data.head()
```

```python
sns.pairplot(amazon_data , diag_kind = "kde")
plt.xlabel("X-Axis")
plt.xticks(rotation=90)
```

```python
sns.distplot(amazon_data['discount percentage'],hist=False)
plt.xticks(rotation=90)
plt.show()

```
