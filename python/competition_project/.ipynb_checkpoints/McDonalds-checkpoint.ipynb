### Which category contributes to the maximum % of Cholesterol in a diet (% daily value)?

```python
df_mcd
```

```python
df_mcd.groupby("Category")["Cholesterol (% Daily Value)"].max()
```

### Inferences 
The maximum Cholestrol present in each category. Breakfast shows to have a food item that has 192 cholestrol. This food item seems to be an outlier in itself and we can be better of by recommending to improve this food item. We can investigate on this food item and report it. (Please refer to extras as to which food in breakfast has highest Cholestrol.)

```python
cholestrol_max = pd.pivot_table(data = df_mcd, values = "Cholesterol (% Daily Value)", index = "Category"  )
cholestrol_max.sort_values(by = "Cholesterol (% Daily Value)", ascending = False)
```

### Inferences
This shows which category correspods to maximum cholestrol in take. Logically Beef and Pork should have been the highest contributor but due to outliers present in breakfast because of which Breakfast shows to be the highest contributor of cholestrol.

```python
plt.figure(figsize = (15,15))
cholestrol_max.sort_values(by = "Cholesterol (% Daily Value)", ascending = False).plot(kind = "bar")
```

### 5. Which item contributes maximum to the Sodium intake?

```python
df_mcd[["Item", "Sodium"]].sort_values(by ="Sodium", ascending = False)
```

```python
sodium_max = pd.pivot_table(data = df_mcd, index = "Item", values = "Sodium")
sodium_max.sort_values(by ="Sodium", ascending = False).head(10)
```

```python
plt.figure(figsize = (15,15))
sodium_max.sort_values(by ="Sodium", ascending = False).head(10).plot(kind = "bar")
```

### Inferences
In the above chart shows the 10 most food items that contain most amount of sodium. 

```python
df_mcd[["Item", "Saturated Fat"]].sort_values(by ="Saturated Fat", ascending = False).head(4)
```

```python
Saturated_Fat_Max = pd.pivot_table(df_mcd, values ="Saturated Fat", index = "Item")
Saturated_Fat = Saturated_Fat_Max.sort_values(by = "Saturated Fat", ascending = False). head(10)
type(Saturated_Fat)
```

```python
#using matplotlib to plot
plt.figure(figsize = (15,15))
Saturated_Fat_Max.sort_values(by = "Saturated Fat", ascending = False). head(10).plot(kind = "bar")
```

```python
#plotting with Seaborn to add more colour to above chart
plt.figure(figsize = (10,10))
sns.barplot(y = Saturated_Fat.index , x = Saturated_Fat["Saturated Fat"])
```

### Inferences

Above shows the top most food item contributing to Saturated Fats.
