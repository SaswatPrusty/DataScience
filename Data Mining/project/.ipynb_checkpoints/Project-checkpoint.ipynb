```python
def treat_outliers(cols):
    IQR = np.percentile(cols, 75)- np.percentile(cols, 25)
    lower_bound = np.percentile(cols, 25) - 1.5*IQR
    upper_bound = np.percentile(cols, 75) + 1.5*IQR
    
    return lower_bound, upper_bound
```

```python
for m in bank_df.columns:
    lower_bound , upper_bound = treat_outliers(bank_df[m])
    bank_df[m] = np.where(bank_df[m]> upper_bound, upper_bound, bank_df[m])
    bank_df[m] = np.where(bank_df[m]< lower_bound, lower_bound, bank_df[m])
    
```

```python
wss = []
for i in range(1,11):
    KM = KMeans(n_clusters=i)
    KM.fit(scaled_df)
    wss.append(KM.inertia_)
```

2.2 Data Split: Split the data into test and train, build classification model CART, Random Forest, Artificial Neural Network


```python
X = insurance_df.drop("Claimed", axis=1)

y = insurance_df.pop("Claimed")

X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, train_labels, test_labels = train_test_split(X, y, test_size=.30, random_state=1)
```

```python
dt_model = DecisionTreeClassifier(criterion = "gini")
dt_model.fit(X_train, train_labels)
```

```python
dt_model.fit(X_train, train_labels)
```

```python
from sklearn import tree

train_char_label = ['No', 'Yes']
ld_Tree_File = open('ld_Tree_File.dot','w')
dot_data = tree.export_graphviz(dt_model, 
                                out_file=ld_Tree_File, 
                                feature_names = list(X_train), 
                                class_names = list(train_char_label))

ld_Tree_File.close()
```

```python
ytrain_predict_CART = dt_model.predict(X_train)
ytest_predict_CART = dt_model.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
dt_model.score(X_test, ytest_predict_CART)
```

```python
rfcl = RandomForestClassifier(n_estimators=300,min_samples_split=4, min_samples_leaf=4)
```

```python
rfcl.fit(X_train, train_labels)
```

```python
ytrain_predict_RF = rfcl.predict(X_train)
ytest_predict_RF = rfcl.predict(X_test)
```

```python
rfcl.score(X_test, ytest_predict_RF)
```

```python
clf = MLPClassifier(hidden_layer_sizes=[300], max_iter=5000,
                     solver='sgd', verbose=True,  random_state=21,tol=0.001)
```

```python
clf.fit(X_train, train_labels)
```

```python
ytest_predict_ANN = clf.predict(X_test)
ytrain_predict_ANN = clf.predict(X_train)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'max_depth': [5,7,10],
    'max_features': [4,6],
    'min_samples_leaf': [5,10],
    'min_samples_split': [50,100],
    'n_estimators': [100,200,300]
}

rfcl = RandomForestClassifier(random_state=1)

grid_search_rfcl = GridSearchCV(estimator = rfcl, param_grid = param_grid, cv = 10)
```

```python
grid_search_rfcl.best_params_
```

```python
best_grid_rfcl = grid_search_rfcl.best_estimator_
```
