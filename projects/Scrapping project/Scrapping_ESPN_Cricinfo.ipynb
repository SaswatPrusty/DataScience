```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://stats.espncricinfo.com/ci/content/records/282827.html"
```

```python
url
```

```python
text = urlopen(url)
```

```python
text
```

```python
soup = BeautifulSoup(text,"lxml")
```

```python
soup
```

```python
table = soup.findAll("table" , attrs = {"class": "engineTable"})
table
```

```python
len(table)
```

```python
type(table[0])
```

```python
tr_list = table[0].findAll("tr", attrs = {"class":"data1"})
```

```python
len(tr_list)
```

```python
tr_list[0]
```

```python
all_players = []
for tr in tr_list:
    td_list = tr.findAll("td")
    row_data = []
    for td in td_list:
        td_str = str(td)
        cleantext = BeautifulSoup(td_str,"lxml").get_text()
        row_data.append(cleantext)
    all_players.append(row_data)    
```

```python
all_players
```

```python
all_players_df = pd.DataFrame(all_players)
```
