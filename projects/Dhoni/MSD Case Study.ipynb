```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
matches_df= pd.read_csv("matches.csv")
matches_df.head()
```

```python
deliveries_df = pd.read_csv("deliveries.csv")
deliveries_df.head()
```

## To analyze batting statistics of MSD and Determine if he is still fit to play

```python
MS_deliveries_df = deliveries_df[deliveries_df.batsman == "MS Dhoni"]
MS_deliveries_df
```

```python
#total number of matches
len(MS_deliveries_df.match_id.unique())
```

```python
MS_matches_df = matches_df[matches_df.id.isin((MS_deliveries_df.match_id.unique()))]
MS_matches_df
```

```python
MS_matches_df.info()
```

```python
MS_matches_df.date = pd.to_datetime(MS_matches_df.date)
MS_matches_df.info()
```

```python
MS_matches_df =MS_matches_df.sort_values("date")
```

```python
MS_matches_df.reset_index(inplace=True, drop = True)
MS_matches_df.head()
```

```python
MS_deliveries_df.head(50)
```

```python
runs_ls = []
for m in MS_matches_df.id:
    temp_df = MS_deliveries_df[MS_deliveries_df.match_id == m]
    runs_scored = sum(temp_df.batsman_runs)
    runs_ls.append(runs_scored)
    #print(runs_scored)
    
```

```python
# runs scored in first 10 balls, 10-30balls, 30+ balls
balls_faced = 0
runs_10b = []
runs_10_30b = []
runs_30_plusb = []
balls_ls = []
for m in MS_matches_df.id:
    temp_df = MS_deliveries_df[MS_deliveries_df.match_id ==m]
    temp_df_wo_wides = temp_df[temp_df.wide_runs == 0]
    balls_faced = len(temp_df_wo_wides)
    balls_ls.append(balls_faced)
    if balls_faced <=10:
        runs_10b.append(sum(temp_df_wo_wides.batsman_runs))
        runs_10_30b.append(0)
        runs_30_plusb.append(0)
    elif (balls_faced > 10) & (balls_faced <=30) :
        runs_10b.append(sum(temp_df_wo_wides.batsman_runs[:10]))
        runs_10_30b.append(sum(temp_df_wo_wides.batsman_runs[10:30]))
        runs_30_plusb.append(0)
    else:
        runs_10b.append(sum(temp_df_wo_wides.batsman_runs[:10]))
        runs_10_30b.append(sum(temp_df_wo_wides.batsman_runs[10:30]))
        runs_30_plusb.append(sum(temp_df_wo_wides.batsman_runs[30:]))



```

```python
MS_matches_df["Runs"] = runs_ls
MS_matches_df["Balls"] = balls_ls
MS_matches_df["Runs_10b"] = runs_10b
MS_matches_df["Runs_10_30b"] = runs_10_30b
MS_matches_df["Runs_30_plusb"] = runs_30_plusb
MS_matches_df.head(10)
```

```python
MS_matches_df[["id", "Season", "Runs", "Runs_10b", "Runs_10_30b", "Runs_30_plusb"]]
```

```python
MS_matches_df.groupby("Season")["Runs"].sum().plot(kind = "bar")
plt.ylabel("Total Runs")
plt.title("MSD Runs by Season")
```

```python
MS_matches_df.groupby("Season")["Runs"].mean().plot(kind = "bar")
plt.ylabel("Average")
plt.title("MSD Average by Season")
```

```python
MS_matches_df.groupby("Season")["Runs_30_plusb"].sum().plot(kind = "bar")
plt.ylabel("Runs")
plt.title("MSD Runs after 30 balls by Season")
```
