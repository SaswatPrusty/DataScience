```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
numeric_columns = ["carat", "table", "x", "y", "z"]
for column in df_zirc[numeric_columns].columns:
    lr,ur=remove_outlier(df_zirc[column])
    df_zirc[column]=np.where(df_zirc[column]>ur,ur,df_zirc[column])
    df_zirc[column]=np.where(df_zirc[column]<lr,lr,df_zirc[column]) 
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
    return(vif_data)
```

```python
calculate_vif(df_zirc.iloc[:,:-1])
```
